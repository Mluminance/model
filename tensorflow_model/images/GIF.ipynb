```python
import os
import imageio
```

```python
img_dir = "/content/drive/MyDrive/Colab Notebooks/Daylight Autonomy/images/generated_img"
```

```python
images = []
for file_name in sorted(os.listdir(img_dir)):
    if file_name.endswith('.png'):
        file_path = os.path.join(img_dir, file_name)
        images.append(imageio.imread(file_path))
imageio.mimsave('generated.gif', images, fps=10)
```
